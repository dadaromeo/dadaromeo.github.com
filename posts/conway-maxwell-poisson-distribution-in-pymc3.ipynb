### Conway-Maxwell Poisson distribution in PyMC3

I recently come discovered a two parameters distribution for heavy-tail count data.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.style.use("bmh")
```

```python
from scipy.stats import norm
```

```python
a = np.linspace(-1,1)

plt.plot(a, norm.pdf(a))
```
